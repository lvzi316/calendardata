```python
import pandas as pd
import json

# Read the JSONL file into a DataFrame
with open('/Users/chaopan/Downloads/Deepseek/Deepseek-r1/2025-03-18-03-40-39.635975/eval_report/metric_results.jsonl', 'r') as f:
    data = [json.loads(line) for line in f]

df = pd.DataFrame(data)


# Show full content in each cell
pd.set_option('display.max_colwidth', None)

# Display the first few rows
display(df.head())

# Basic analysis
print("Number of rows:", len(df))
print("Columns:", df.columns.tolist())
print("Missing values per column:\n", df.isnull().sum())
print("Data types:\n", df.dtypes)
print("Summary statistics (numeric columns):\n", df.describe())
```

```python
import json
import re

# Load the file
with open("single_sample.json", "r") as f:
    raw = f.read()

# Remove escaped newlines and extra backslashes
cleaned = raw.replace('\\n', '\n').replace('\\"', '"')

# Remove comments if present (lines starting with //)
cleaned = re.sub(r'//.*\n', '', cleaned)

# Parse the cleaned JSON
data = json.loads(cleaned)

# Pretty print the JSON
from pprint import pprint
pprint(data)
```
